```python
import requests
from bs4 import BeautifulSoup

def check_website_scrapability(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for error HTTP statuses
        soup = BeautifulSoup(response.text, 'html.parser')

        # Check for common HTML elements
        if soup.find('title') and soup.find('body'):
            print("Website structure seems suitable for scraping.")
        else:
            print("Website structure might be complex or dynamic.")

        # Check for robots.txt
        robots_url = f"{url}/robots.txt"
        robots_response = requests.get(robots_url)
        robots_soup = BeautifulSoup(robots_response.text, 'html.parser')

        if robots_soup.find('Disallow'):
            disallow_rules = robots_soup.find_all('Disallow')
            for rule in disallow_rules:
                if rule.text.strip() == '/':
                    print("Website explicitly disallows scraping.")
                    return

        print("No explicit disallow rule found in robots.txt.")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching the website: {e}")

if __name__ == "__main__":
    url = "https://www.der-rollenshop.de/Inlineskates"
    check_website_scrapability(url)
```

```python
page = requests.get("https://www.der-rollenshop.de/Urbanskates-Freeskates") 
source = page.content
print (source)
```

```python
soup = BeautifulSoup (source , 'lxml')
print (soup)
```

```python
Skate_Name = []
Skate_Price = []
Skate_Brand = []

name1 = soup.find_all("a", {"class": "text-clamp-2"})

for name in name1:
    print(name.text)
    
```

```python
prices = soup.find_all("div", {"class": "price productbox-price"})

for price in prices:
    price_text = price.text.strip()
    cleaned_price = price_text.replace("$0", "").replace('"', "").replace("*", "").replace("€", "").strip()

    print(cleaned_price)
```

```python
brands = soup.find_all("a", {"class": "text-clamp-2"})

for brand in brands:
    brand_text = brand.text.strip()  
    first_word = brand_text.split()[0]  
    print(first_word)
```

```python
Skate_Name = []
Skate_Price = []
Skate_Brand = []


for i in range(len(prices)):
    Skate_Name.append(name1[i].text)
    Skate_Price.append(prices[i].text.replace("$0", "").replace('"', "").replace("*", "").replace("€", "").strip())
    Skate_Brand.append(brands[i].text.split()[0])
    print(Skate_Name[i] )
    print(Skate_Brand[i])
    print(Skate_Price[i])
    
```

```python
file_list=[Skate_Name , Skate_Brand, Skate_Price]
ex = zip_longest(*file_list)
```

```python

with open("skate list.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Skate Name","Skate Price", "Skate Brand"])

    for row in ex:
        writer.writerow({row[0] , row[1] , row[2] })  # Write the entire row as a list
        
print("data saved ")
```
